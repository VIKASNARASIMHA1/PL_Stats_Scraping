# Sponsors in Premier League

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
x = requests.get('https://www.myfootballfacts.com/premier-league/all-time-premier-league/premier_league_shirt_sponsors/')
parsedData = BeautifulSoup(x.text, "html.parser")
container = (parsedData.findAll("table"))[2]
conti = container.find("tbody") 
data = conti.findAll("tr")
j = len(data)
list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
list6 = []
list7 = []
list8 = []
list9 = []
list10 = []
list11 = []
list12 = []
list13 = []
for i in range(1,j+1):
    data1 = (container.findAll("td",{"class","dtr{} dtc1".format(i)}))
    data2 = (container.findAll("td",{"class","dtr{} dtc2".format(i)}))
    data2 = [cell if cell.get_text(strip=True) else "-" for cell in data2]
    data3 = (container.findAll("td",{"class","dtr{} dtc3".format(i)}))
    data3 = [cell if cell.get_text(strip=True) else "-" for cell in data3]
    data4 = (container.findAll("td",{"class","dtr{} dtc4".format(i)}))
    data4 = [cell if cell.get_text(strip=True) else "-" for cell in data4]
    data5 = (container.findAll("td",{"class","dtr{} dtc5".format(i)}))
    data5 = [cell if cell.get_text(strip=True) else "-" for cell in data5]
    data6 = (container.findAll("td",{"class","dtr{} dtc6".format(i)}))
    data6 = [cell if cell.get_text(strip=True) else "-" for cell in data6]
    data7 = (container.findAll("td",{"class","dtr{} dtc7".format(i)}))
    data7 = [cell if cell.get_text(strip=True) else "-" for cell in data7]
    data8 = (container.findAll("td",{"class","dtr{} dtc8".format(i)}))
    data8 = [cell if cell.get_text(strip=True) else "-" for cell in data8]
    data9 = (container.findAll("td",{"class","dtr{} dtc9".format(i)}))
    data9 = [cell if cell.get_text(strip=True) else "-" for cell in data9]
    data10 = (container.findAll("td",{"class","dtr{} dtc10".format(i)}))
    data10 = [cell if cell.get_text(strip=True) else "-" for cell in data10]
    data11 = (container.findAll("td",{"class","dtr{} dtc11".format(i)}))
    data11 = [cell if cell.get_text(strip=True) else "-" for cell in data11]
    data12 = (container.findAll("td",{"class","dtr{} dtc12".format(i)}))
    data12 = [cell if cell.get_text(strip=True) else "-" for cell in data12]
    data13 = (container.findAll("td",{"class","dtr{} dtc13".format(i)}))
    data13 = [cell if cell.get_text(strip=True) else "-" for cell in data13]
    for j in data1:
        list1 += j
    for j in data2:
        list2 += j
    for j in data3:
        list3 += j
    for j in data4:
        list4 += j
    for j in data5:
        list5 += j
    for j in data6:
        list6 += j
    for j in data7:
        list7 += j
    for j in data8:
        list8 += j
    for j in data9:
        list9 += j
    for j in data10:
        list10 += j
    for j in data11:
        list11 += j
    for j in data12:
        list12 += j
    for j in data13:
        list13 += j
list2.pop()
list3.pop()
list4.pop()
list5.pop()
list6.pop()
list7.pop()
list8.pop()
list9.pop()
list10.pop()
list11.pop()
list12.pop()
list13.pop()
df = pd.DataFrame({'Sl_Number':df.index+1,'Club':list1 , '12':list2, '13':list3, '14':list4, '15':list5, '16':list6,'17':list7,'18':list8,'19':list9,'20':list10,'21':list11,'22':list12,'23':list13}) 
df.to_csv('SponsorPL.csv', encoding='utf-8-sig', index = False)
```
