# Winners of each Premier League Season

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
x = requests.get('https://www.myfootballfacts.com/premier-league-winners/')
parsedData = BeautifulSoup(x.text, "html.parser")
container = parsedData.find("table")
conti = container.find("tbody") 
data = conti.findAll("tr")
j = len(data)
list1 = []
list2 = []
list3 = []
list4 = []
list5 = []
list6 = []
list7 = []
list8 = []
list9 = []
list10 = []
list11 = []
list12 = []
list13 = []
list14 = []
list15 = []
for i in range(j+1):
    data1 = (container.findAll("td",{"class","dtr{} dtc0".format(i)}))
    data2 = (container.findAll("td",{"class","dtr{} dtc1".format(i)}))
    data3 = (container.findAll("td",{"class","dtr{} dtc2".format(i)}))
    data4 = (container.findAll("td",{"class","dtr{} dtc3".format(i)}))
    data5 = (container.findAll("td",{"class","dtr{} dtc4".format(i)}))
    data5 = [cell if cell.get_text(strip=True) else "-" for cell in data5]
    data6 = (container.findAll("td",{"class","dtr{} dtc5".format(i)}))
    data7 = (container.findAll("td",{"class","dtr{} dtc6".format(i)}))
    data8 = (container.findAll("td",{"class","dtr{} dtc7".format(i)}))
    data9 = (container.findAll("td",{"class","dtr{} dtc8".format(i)}))
    data10 = (container.findAll("td",{"class","dtr{} dtc9".format(i)}))
    data11 = (container.findAll("td",{"class","dtr{} dtc10".format(i)}))
    data12 = (container.findAll("td",{"class","dtr{} dtc11".format(i)}))
    data13 = (container.findAll("td",{"class","dtr{} dtc12".format(i)}))
    for j in data1:
        list1 += j
    for j in data2:
        list3 += j
    for j in data3:
        list5 += j
    for j in data4:
        list6 += j
    for j in data5:
        list7 += j
    for j in data6:
        list8 += j
    for j in data7:
        list9 += j
    for j in data8:
        list10 += j
    for j in data9:
        list11 += j
    for j in data10:
        list12 += j
    for j in data11:
        list13 += j
    for j in data12:
        list14 += j
    for j in data13:
        list15 += j
for k in list1:
    list2 += k
for k in list3:
    list4 += k
list7.pop()
list7 = list7 + [None] * (len(list2) - len(list7))
df = pd.DataFrame({'SL_Number':df.index+1,'Season':list2 , 'Winner':list4, 'Played':list5,'Won':list6, 'Draw':list7, 'Lost': list8, 'GF':list9, 'GA':list10, 'GD':list11, 'Points':list12, 'PPG':list13, 'Manager':list14, 'Captain': list15}) 
df.to_csv('WinnerDetailsPL.csv', index=False, encoding='utf-8-sig')  
```
